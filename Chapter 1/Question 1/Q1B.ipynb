```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/Data/data.csv')
df
```

```python
null_counts = df.isnull().sum()
print("Number of null values in each column:")
print(null_counts)
```

```python
df['z'] = df['z'].replace('a', '1')
df['z'] = df['z'].replace('b', '0')
df
```

```python
X = df[['x','y']]
X1 = df[['x']]
X2 = df[['y']]
y = df[['z']]
print(X1.shape)
print(X2.shape)
print(y.shape)
```

```python
plt.scatter(X1, X2, c=y.values)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = LogisticRegression(solver='sag', max_iter=200, random_state=76)
model.fit(x_train, y_train)
model.predict(x_test), y_test
```

```python
model.predict_proba(x_test), y_test
```

```python
model.predict_log_proba(x_test), y_test
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model1 = SGDClassifier(loss='log_loss', random_state=76)
model1.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
x1_min, x2_min = X.min(0)
x1_max, x2_max = X.max(0)
```

```python
n = 500
x1r = np.linspace(x1_min, x1_max, n)
x2r = np.linspace(x2_min, x2_max, n)
x1m, x2m = np.meshgrid(x1r, x2r)
```

```python
Xm = np.stack((x1m.flatten(), x2m.flatten()), axis=1)
ym = model.decision_function(Xm)

# Assign colors to classes (0 for blue, 1 for red)
colors = np.array(['blue', 'red'])

plt.scatter(X1, X2, c=y.values)
plt.contour(x1m, x2m, ym.reshape(x1m.shape), levels=[-1,0,1])
plt.show()
```

```python
Xm = np.stack((x1m.flatten(), x2m.flatten()), axis=1)
ym = model1.decision_function(Xm)

# Assign colors to classes (0 for blue, 1 for red)
colors = np.array(['blue', 'red'])

plt.scatter(X1, X2, c=y.values)
plt.contour(x1m, x2m, ym.reshape(x1m.shape), levels=[-1,0,1])
plt.show()
```
