```python
"""Generate Dataset"""
import matplotlib.pyplot as plt
from sklearn.datasets import  make_classification
X, y = make_classification(n_samples=1000, n_features=3,
                           n_redundant=0, n_clusters_per_class=1,
                           class_sep=1, n_classes=4, random_state=76)

print(X.shape, y.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d", elev=-150, azim=110)
ax.scatter(
    X[:, 0],
    X[:, 1],
    X[:, 2],
    c=y
)

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
```

```python
Xm, ym = make_classification(n_samples=1000, n_features=3,
                           n_redundant=0, n_clusters_per_class=1,
                           class_sep=0.1, n_classes=4, random_state=76)

print(Xm.shape, ym.shape)
fig = plt.figure()
ax = fig.add_subplot(projection="3d", elev=-150, azim=110)
ax.scatter(
    Xm[:, 0],
    Xm[:, 1],
    Xm[:, 2],
    c=ym
)

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
```
