```python
"""Generate Dataset"""
import matplotlib.pyplot as plt
from sklearn.datasets import  make_classification
X, y = make_classification(n_samples=1000, n_features=3,
                           n_redundant=0, n_clusters_per_class=1,
                           class_sep=1, n_classes=4, random_state=76)

print(X.shape, y.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d", elev=-150, azim=110)
ax.scatter(
    X[:, 0],
    X[:, 1],
    X[:, 2],
    c=y
)

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
```

```python
Xm, ym = make_classification(n_samples=1000, n_features=3,
                           n_redundant=0, n_clusters_per_class=1,
                           class_sep=0.1, n_classes=4, random_state=76)

print(Xm.shape, ym.shape)
fig = plt.figure()
ax = fig.add_subplot(projection="3d", elev=-150, azim=110)
ax.scatter(
    Xm[:, 0],
    Xm[:, 1],
    Xm[:, 2],
    c=ym
)

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = LogisticRegression(multi_class='multinomial',solver='sag', max_iter=200, random_state=76)
model.fit(x_train, y_train)
model.predict(x_test), y_test
```

```python
model.predict_proba(x_test), y_test
```

```python
model.predict_log_proba(x_test), y_test
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model1 = LogisticRegression(multi_class='ovr',solver='sag', max_iter=200, random_state=76)
model1.fit(x_train, y_train)
model1.predict(x_test), y_test
```

```python
model1.predict_proba(x_test), y_test
```

```python
model1.predict_log_proba(x_test), y_test
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
model2 = SGDClassifier(loss='log_loss', random_state=76)
model2.fit(x_train, y_train)
```

```python
model2.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
import numpy as np
```

```python
# Generate mesh grid
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
z_min, z_max = X[:, 2].min() - 0.1, X[:, 2].max() + 0.1
xx, yy, zz = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100),
                         np.linspace(z_min, z_max, 100))
```

```python
# Make predictions
mesh_input = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
predictions = model.predict(mesh_input)
predictions = np.round(predictions).astype(int)

# Plot decision regions
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')  # Original data points
ax.scatter(mesh_input[:, 0], mesh_input[:, 1], mesh_input[:, 2], c=predictions, cmap='viridis', alpha=0.1)  # Decision regions
plt.show()
```

```python
predictions1 = model1.predict(mesh_input)
predictions1 = np.round(predictions1).astype(int)

# Plot decision regions
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')  # Original data points
ax.scatter(mesh_input[:, 0], mesh_input[:, 1], mesh_input[:, 2], c=predictions1, cmap='viridis', alpha=0.1)  # Decision regions
plt.show()
```
